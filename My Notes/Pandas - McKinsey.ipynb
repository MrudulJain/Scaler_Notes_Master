```python
#filter out risky movies - movies whose budget is greater than avg. revenue of director

def risky(x):
    #print(x.shape, "\n")
    x["risky"] = (x["budget"] - x["revenue"].mean()) > 0
    return x

risky_movies = data.groupby("director_name").apply(risky)
risky_movies
```

```python
#Find squikest directors --> most movies/year
aggr_data = data.groupby("director_name")[["title", "year"]].agg({"year": ["min", "max"], "title":["count"]})
aggr_data
```

```python
aggr_data.columns
```

```python
aggr_data.columns = ["_".join(i) for i in aggr_data.columns]
aggr_data.reset_index(inplace=True)
aggr_data
```

```python
aggr_data["years_active"] = aggr_data["year_max"] - aggr_data["year_min"]
aggr_data["movies_per_year"] = aggr_data["title_count"]/aggr_data["years_active"]
aggr_data.sort_values("movies_per_year", ascending=False)
```
